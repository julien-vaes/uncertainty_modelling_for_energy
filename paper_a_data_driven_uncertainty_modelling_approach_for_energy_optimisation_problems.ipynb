## Details on the groups of the attributes

```python
# the blocs definition
theColsBlocs = {}

# Elec
theColsBlocs['Elec'] = ['Elec']

# Gas demand for the regions: ['EA', 'EM', 'NE', 'NO', 'NT', 'NW', 'SC', 'SE', 'SO', 'SW','WM']
theRegionsPart1 = ['EA', 'EM', 'NE', 'NO', 'NT', 'NW', 'SC', 'SE', 'SO', 'SW','WM']
theColsBlocs['GasDemPart1'] = ['GasDem_'+s for s in theRegionsPart1]

# Gas demand for the regions: ['WN','WS']
theRegionsPart2 = ['WN','WS']
theColsBlocs['GasDemPart2'] = ['GasDem_'+s for s in theRegionsPart2]

# Gas demand for all regions
theColsBlocs['GasDem'] = theColsBlocs['GasDemPart1'] # + theColsBlocs['GasDemPart2']

# 'Solar'
theColsBlocs['Solar'] = ['Solar_EA', 'Solar_EM', 'Solar_NE', 'Solar_NO','Solar_NT', 'Solar_NW', 'Solar_SC', 'Solar_SE', 'Solar_SO', 'Solar_SW', 'Solar_WM', 'Solar_WN', 'Solar_WS']

#'Temp'
theColsBlocs['Temp'] = ['Temp_EA', 'Temp_EM', 'Temp_NE', 'Temp_NO', 'Temp_NT', 'Temp_NW', 'Temp_SC', 'Temp_SE', 'Temp_SO', 'Temp_SW', 'Temp_WM', 'Temp_WN', 'Temp_WS']

# 'WindOff'
theColsBlocs['WindOff'] = ['WindOff_EA', 'WindOff_EM', 'WindOff_NE', 'WindOff_NO', 'WindOff_NT', 'WindOff_NW', 'WindOff_SC', 'WindOff_SE', 'WindOff_SO', 'WindOff_SW', 'WindOff_WN', 'WindOff_WS']

# 'WindOn'
theColsBlocs['WindOn'] = ['WindOn_EA', 'WindOn_EM', 'WindOn_NE', 'WindOn_NO', 'WindOn_NT', 'WindOn_NW', 'WindOn_SC', 'WindOn_SE', 'WindOn_SO', 'WindOn_SW', 'WindOn_WM', 'WindOn_WN', 'WindOn_WS']

# # 'Interconnector'
# theColsBlocs['Interconnector'] = ['Int_BRITNED','Int_FABLINK','Int_MOYLE','Int_NEMO','Int_Northconnet','Int_VikingLink']
```

## Details on which attributes must be clustered together

```python
# attributes to consider when deriving the clusters
theDictClustersInfo = {}

# the number of clusters generated
theNClusters = 12

# Cluster on all the attributes
theCluster_All = {}
# theCluster_All['Attributes_info'] = ['Elec','GasDemPart1','GasDemPart2','Interconnector','Solar','Temp','WindOff','WindOn']
theCluster_All['Attributes_info'] = ['Elec','GasDem','Solar','Temp','WindOff','WindOn']
theCluster_All['Output_folder']   = theFolderOutputImages + 'cluster_all/'
theCluster_All['N_Clusters']      = theNClusters
theDictClustersInfo['All']        = theCluster_All

# Cluster on elec and gas attributes
theCluster_ElecGas = {}
theCluster_ElecGas['Attributes_info'] = ['Elec','GasDem']
theCluster_ElecGas['Output_folder']   = theFolderOutputImages + 'cluster_elec_gas/'
theCluster_ElecGas['N_Clusters']      = theNClusters
theDictClustersInfo['ElecGas']        = theCluster_ElecGas

# Cluster on elec, gas 1 and temp as they are highly correlated attributes
theCluster_ElecGasTemp = {}
theCluster_ElecGasTemp['Attributes_info'] = ['Elec','GasDem','Temp']
theCluster_ElecGasTemp['Output_folder']   = theFolderOutputImages + 'cluster_elec_gas_temp/'
theCluster_ElecGasTemp['N_Clusters']      = theNClusters
theDictClustersInfo['ElecGasTemp']        = theCluster_ElecGasTemp

# Cluster on elec, gas 1 and temp as they are highly correlated attributes
theCluster_Seas = {}
theCluster_Seas['Attributes_info'] = ['Elec','GasDem','Temp']
theCluster_Seas['Output_folder']   = theFolderOutputImages + 'cluster_seasonality/'
theCluster_Seas['N_Clusters']      = theNClusters
theDictClustersInfo['Seasonality'] = theCluster_Seas

# Cluster on wind on and off attributes
theCluster_Wind = {}
theCluster_Wind['Attributes_info'] = ['WindOff','WindOn']
theCluster_Wind['Output_folder']   = theFolderOutputImages + 'cluster_wind/'
theCluster_Wind['N_Clusters']      = theNClusters
theDictClustersInfo['Wind']        = theCluster_Wind

# Cluster on solar attributes
theCluster_Solar = {}
theCluster_Solar['Attributes_info'] = ['Solar']
theCluster_Solar['Output_folder']   = theFolderOutputImages + 'cluster_solar/'
theCluster_Solar['N_Clusters']      = theNClusters
theDictClustersInfo['Solar']        = theCluster_Solar

# # Cluster on interconnectors attributes
# theCluster_Int = {}
# theCluster_Int['Attributes_info'] = ['Interconnector']
# theCluster_Int['Output_folder']   = 'outputs/images/cluster_int/'
# theCluster_Int['N_Clusters']      = theNClusters
# theDictClustersInfo['Int']        = theCluster_Int

# Cluster on interconnectors attributes
theCluster_All_No_Int = {}
theCluster_All_No_Int['Attributes_info'] = ['Elec','GasDemPart1','GasDemPart2','Temp','Solar','WindOff','WindOn']
theCluster_All_No_Int['Output_folder'] = theFolderOutputImages + 'outputs/images/cluster_all_not_int/'
theCluster_All_No_Int['N_Clusters'] = theNClusters
theDictClustersInfo['All_No_Int'] = theCluster_All_No_Int

# gets all the attributes to cluster on, e.g. 'Solar' = ['Solar_EA', etc]
for key_loc in theDictClustersInfo:
    dict_loc = theDictClustersInfo[key_loc]
    myClusterAttributes = []
    for attr_loc in dict_loc['Attributes_info']:
        if attr_loc in theColsBlocs.keys():
            myClusterAttributes += theColsBlocs[attr_loc]
        else:
            myClusterAttributes += [attr_loc]
    dict_loc['Attributes'] = myClusterAttributes
```

# Cartesian product sub-clusters

```python
theProbThreshold = 0.02
```

## Count the number of days in each cluster = Cartesian product of sub-clusters

```python
theNDaysCartesianProdClusters = np.zeros((len(theClusterAttribution['Seasonality']),len(theClusterAttribution['Solar']),len(theClusterAttribution['Wind'])),dtype=object)

for d in range(365):
    clust_seas  = -1.0
    clust_solar = -1.0
    clust_wind  = -1.0
    
    for i in range(len(theClusterAttribution['Seasonality'])):
        if d in theClusterAttribution['Seasonality'][i]:
            clust_seas = i
    
    for i in range(len(theClusterAttribution['Solar'])):
        if d in theClusterAttribution['Solar'][i]:
            clust_solar = i
    
    for i in range(len(theClusterAttribution['Wind'])):
        if d in theClusterAttribution['Wind'][i]:
            clust_wind = i
    
    theNDaysCartesianProdClusters[clust_seas,clust_solar,clust_wind] += 1

# removes scenarios with prob < theProbThreshold
for i in range(len(theClusterAttribution['Seasonality'])):
    for j in range(len(theClusterAttribution['Solar'])):
        for k in range(len(theClusterAttribution['Wind'])):
            if theNDaysCartesianProdClusters[i,j,k]/365.0 < theProbThreshold:
                theNDaysCartesianProdClusters[i,j,k] = 0  

# derive prob of each retained scenario
theProbCartesianProdClusters = theNDaysCartesianProdClusters / np.sum(theNDaysCartesianProdClusters)  

# round values
for i in range(len(theClusterAttribution['Seasonality'])):
    for j in range(len(theClusterAttribution['Solar'])):
        for k in range(len(theClusterAttribution['Wind'])):
            theProbCartesianProdClusters[i,j,k] = round(theProbCartesianProdClusters[i,j,k],3)
            
theProbCartesianProdClusters
```
